```python
import yfinance as yf
import pandas as pd
```

```python
TSLA = yf.Ticker("TSLA")
TM = yf.Ticker("TM")
F = yf.Ticker("F")
GM = yf.Ticker("GM")
RACE = yf.Ticker("RACE")
HMC = yf.Ticker("HMC")
XPEV = yf.Ticker("XPEV")
DDAIF = yf.Ticker("DDAIF")
VWAGY = yf.Ticker("VWAGY")
HYMTF = yf.Ticker("HYMTF")
```

```python
tsla_historical = TSLA.history(start="2016-02-28", end="2021-07-28", interval="1d")
tm_historical = TM.history(start="2016-02-28", end="2021-07-28", interval="1d")
f_historical = F.history(start="2016-02-28", end="2021-07-28", interval="1d")
gm_historical = GM.history(start="2016-02-28", end="2021-07-28", interval="1d")
race_historical = RACE.history(start="2016-02-28", end="2021-07-28", interval="1d")
hmc_historical = HMC.history(start="2016-02-28", end="2021-07-28", interval="1d")
xpev_historical = XPEV.history(start="2016-02-28", end="2021-07-28", interval="1d")
ddaif_historical = DDAIF.history(start="2016-02-28", end="2021-07-28", interval="1d")
vwagy_historical = VWAGY.history(start="2016-02-28", end="2021-07-28", interval="1d")
hymtf_historical = HYMTF.history(start="2016-02-28", end="2021-07-28", interval="1d")
```
